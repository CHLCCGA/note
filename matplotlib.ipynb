```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,1000)
#中文标题
plt.rcParams['font.family'] = ['Microsoft YaHei']

plt.figure(figsize=(8,5))
plt.plot(x,np.sin(x),':g',label='sin(x)')
plt.plot(x,np.cos(x),'-.r',label='cos(x)')
plt.legend()
plt.title('title here 标题')
plt.ylabel('ylabelY轴')
plt.show()
```

## bar

```python
cate = ['Apple', 'Huawei', 'Xiaomi', 'Oppo', 'Vivo', 'Meizu']
data1 = [123, 153, 157, 107, 98, 142]
data2 = [56, 77, 93, 68, 45, 67]

plt.figure(figsize=(5,4))
plt.bar(cate,data1,label='Total')
plt.bar(cate,data2,label='Online')
plt.legend()
plt.title('Mobile Phone',loc='right')
plt.show()
```

```python
cate = ['Apple', 'Huawei', 'Xiaomi', 'Oppo', 'Vivo', 'Meizu']
data1 = [123, 153, 157, 107, 98, 142]
data2 = [56, 77, 93, 68, 45, 67]

x_location = np.arange(len(cate))

plt.figure(figsize=(5,4))
plt.bar(x_location,data1,label='ratail',width=0.3)
plt.bar(x_location+0.3,data2,label='online',width=0.3)

plt.xticks(x_location+0.15,cate)
plt.legend()
plt.title('mobile phone'
          ,loc='left'
          ,fontdict={'fontsize':18,'color':'blue'})
plt.show()
```

## barh

```python
cate = ['Apple', 'Huawei', 'Xiaomi', 'Oppo', 'Vivo', 'Meizu']
data1 = [123, 153, 157, 107, 98, 142]
data2 = [56, 77, 93, 68, 45, 67]

x_location = np.arange(len(cate))

# barh yticks
plt.figure(figsize=(5,3))
plt.barh(x_location,data1,label='retail',height=0.3)
plt.barh(x_location+0.3,data2,label='online',height=0.3)

plt.yticks(x_location+0.15,cate)
plt.legend()
plt.title('mobile phone'
          ,loc='center'
          ,fontdict={'fontsize':18,'color':'red'})
plt.show()
```

## pie

```python
cate = ['Apple', 'Huawei', 'Xiaomi', 'Oppo', 'Vivo', 'Meizu']
data = [153, 124, 107, 99, 89, 46]

plt.figure(figsize=(5,5))
plt.pie(data,labels=cate
        ,explode=[0,0,0,0,0.2,0]
        ,autopct='%.2f%%'
        ,shadow=True
        ,colors=['#FFC0CB', '#FF00FF', '#FFD700', '#FFA500', '#00FF7F', 'blue']
        ,textprops={'fontsize':13,'color':'black'})
plt.legend()
plt.show()
```

## dynamic

```python
from IPython import display
```

```python
x, y1, y2 = [], [], []
plt.ion()    # 交互模式打开
for i in np.linspace(0,20,500):
    x.append(i)
    y1.append(np.cos(i))
    y2.append(np.sin(i))
    plt.figure(figsize=(12, 2))
    plt.clf()   
    plt.ylim(-1, 1)    # 设置y轴刻度值最大最小值
    plt.plot(x, y1, '--r')
    plt.plot(x, y2, 'g')
    plt.pause(0.1)
    display.clear_output(wait=True)    # 清除输出
plt.ioff()   # 关闭交互模式
```

```python
# 可用风格
style_list = plt.style.available

# 数据
cate = ['Apple', 'Huawei', 'Xiaomi', 'Oppo', 'Vivo', 'Meizu']
data1 = [89, 45, 56, 74, 23, 21]
data2 = [56, 77, 93, 68, 45, 67]
x_location = np.arange(len(cate))

plt.figure(figsize=(20, 40))
for i, style in enumerate(style_list):
    
    with plt.style.context(style):  # 设置图表风格     
        plt.subplot(7, 4, i+1)
        plt.bar(x_location, data1, label='线下', width=0.3)
        plt.bar(x_location+0.3, data2, label='线上', width=0.3)

        # 为X轴设置刻度值
        plt.xticks(x_location+0.15, cate)
        plt.title(style,
                 fontdict={'fontsize':20,'color':'red'})
plt.show()
```
